# Clean DATA for dataset 

Dataset provided by Kushagra Data collation team 

```python
import pandas as pd
import numpy as np
```

```python
dfMSA=pd.read_csv('title_mor.csv')
dfMSA.head()
```

# Cleaning MSA

```python
dfMSA=dfMSA.drop(columns=['Case','lemma', 'Gender','Number','State','Gloss','Proclitic 1','Proclitic 2','Proclitic 0','Aspect','Mood','Person','Voice','Enclitic 0'])
dfMSA=dfMSA.drop_duplicates()
#dfMSA=dfMSA.dropna()
dfMSA["POS"] = dfMSA["POS"].apply(lambda x: x.replace("Proper Noun", "ProperNoun"))
```

```python
dfMSA
```

```python
copofpos=dfMSA.groupby('sentence_id')['POS'].apply(' '.join).reset_index()
```

```python
copofword=dfMSA.groupby('sentence_id')['word'].apply(' '.join).reset_index()
```

```python
copofpos
```

```python
copofword
```

```python
df=pd.merge(copofword,copofpos)
```

# Save files as CSV and Merge them 

```python
df.to_csv('Sentences-POS 7460.csv', index=False)
```
